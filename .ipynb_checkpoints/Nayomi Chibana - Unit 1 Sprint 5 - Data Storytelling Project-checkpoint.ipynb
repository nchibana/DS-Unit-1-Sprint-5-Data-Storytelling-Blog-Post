```python
import pandas as pd
```

```python
df = pd.read_csv('/users/nchib/desktop/ECR-data-1March2018/ECR-data-1March2018/data.csv')
print(df.shape)
df.head()
```

```python
df.describe()
```

```python
df_scores = df.copy()
df_scores[['Q3','Q15','Q19','Q22','Q25','Q27','Q29','Q31','Q33','Q35']] = df_scores[['Q3','Q15','Q19','Q22','Q25','Q27','Q29','Q31','Q33','Q35']].apply(lambda x: 6-x, axis=1)
df_scores.head()
```

```python
df_scores['avoidance_score'] = (df_scores[df_scores.columns[0:35:2]].sum(axis=1))/18
df_scores.head()
```

```python
df_scores['anxiety_score'] = (df_scores[df_scores.columns[1:36:2]].sum(axis=1))/18
df_scores.head()
```
