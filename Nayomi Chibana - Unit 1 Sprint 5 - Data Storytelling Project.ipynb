```python
df_score = df.copy()
reverse = ['Q3','Q15','Q19','Q22','Q25','Q27','Q29','Q31','Q33','Q35']
df_score[reverse] = df_score[reverse].applymap(lambda x: 6-x if x > 0 else x)
df_score.head()
```

```python
df_score['avoidance_score'] = (df_score[df_score.columns[0:35:2]].sum(axis=1))/18
df_score.head()
```

```python
df_score['anxiety_score'] = (df_score[df_score.columns[1:36:2]].sum(axis=1))/18
df_score.head()
```

```python
df_score.drop(df_score[(df_score['avoidance_score'] == 0) | (df_score['anxiety_score'] == 0)].index, inplace = True)
df_score.describe()
```

```python
df_score.drop(df_score[(df_score['age'] > 100) | (df_score['gender'] == 0)].index, inplace = True)
df_score.describe()
```
