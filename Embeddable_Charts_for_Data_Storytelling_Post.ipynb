#Analysis of Stress Scores by Religion

```
from scipy import stats
```

```
protest = df[df.religion == 6]
mormon = df[df.religion == 5]
```

```
stats.ttest_ind(mormon['stress_score'], protest['stress_score'], nan_policy='omit')
```

The outcome of this t-test indicates that we should reject the null hypothesis that the mean of stress scores for Mormon survey respondents is equal to the mean of stress scores for Protestant survey respondents. 
We would conclude due to the positive t-statistic here that the self-reported stress scores of Mormons are significantly higher than those of Protestants.

```
buddhist = df[df.religion == 3]
atheist = df[df.religion == 2]
```

```
stats.ttest_ind(mormon['stress_score'], buddhist['stress_score'], nan_policy='omit')
```
